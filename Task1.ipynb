3. Write a Python program to count the number of strings from a given list of strings. The string length is 2 or more and the first and last characters are the same.

```
Sample List : ['abc', 'xyz', 'aba', '1221']
Expected Result : 2
```

````python
mylist=['abc', 'xyz', 'aba', '1221']
result=[]
for i in mylist:
  if len(i)>2 and i[0]==i[-1]:
    result.append(i)
print(len(result))
````

11.Write a Python script to sort (ascending and descending) a dictionary by value


````python
import operator
mydict = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
sorted_dict = sorted(mydict.items(), key=operator.itemgetter(1))
print('Dictionary in ascending order by value: ',sorted_dict)
````

````python
import operator
mydict = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
sorted_dict = dict( sorted(mydict.items(), key=operator.itemgetter(1), reverse=True))
print('Dictionary in descending order by value: ',sorted_dict)
````

14.Write a Python program to create a dictionary with a number and its corresponding square from 1 to input number. And also check if the input number is less than 10

```
Eg:
Input : 3
Output : {1:1, 2:4, 3:9}
```

````python
num=int(input('Enter the number: '))
if(num<10):
    print({i:i*i for i in range(1,num+1)})
else:
  print("Input number should be less than 10")
````

17. Write a Python program to get the top three items in a shop.

```
Sample data: {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5': 24}

Expected Output:
item4 55
item1 45.5
item3 41.3
```


````python

import operator
mydict = {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5': 24}
sorted_dict = dict( sorted(mydict.items(), key=operator.itemgetter(1), reverse=True))
print('Items in descending order by value: ',sorted_dict)
res = dict(list(sorted_dict.items())[0: 3])
print("Top three items are", res)
````
